## Show Duplicated Rows

```python
# Step 1: Get all column names dynamically
query_get_columns = """
    SELECT GROUP_CONCAT(COLUMN_NAME)
    FROM INFORMATION_SCHEMA.COLUMNS
    WHERE TABLE_NAME = 'sample_month_01' AND TABLE_SCHEMA = 'uber';
"""

with engine.connect() as connection:
    columns = connection.execute(query_get_columns).fetchone()[0]  # Get column list as string
```

```python
columns
```

```python
', '.join([f'{col}' for col in columns.split(',')])
```

```python
#dupliactes query
dup_query= f"""
            select {', '.join([f'{col}' for col in columns.split(',')])} , count(*)
            from sample_month_01
            GROUP BY {', '.join([f'{col}' for col in columns.split(',')])}
            HAVING COUNT(*) > 1;
"""
pd.read_sql(dup_query,engine)
```

```python
#to confirm this result is true

#show data of month 1
query="""select * from sample_month_01
        order by 'index' asc;
"""
# Display all columns
pd.set_option("display.max_columns", None)

sample_month_1=pd.read_sql(query,engine)
```

## Rename Value

```python
# in data describe from source in Hvfhs_license_num column 
# HV0002: Juno
# HV0003: Uber
# HV0004: Via
# HV0005: Lyft
from time import time
from sqlalchemy import text
t_start=time()

re_query = text("""
    UPDATE sample_month_01
    SET Hvfhs_license_num = CASE
                                WHEN Hvfhs_license_num = 'HV0002' THEN 'Juno'
                                WHEN Hvfhs_license_num = 'HV0003' THEN 'Uber'
                                WHEN Hvfhs_license_num = 'HV0004' THEN 'Via'
                                WHEN Hvfhs_license_num = 'HV0005' THEN 'Lyft'
                                ELSE Hvfhs_license_num  -- Keep the existing value if no match
    END;
""")
with engine.connect() as conn:
    conn.execute(re_query)
    
t_end=time()
print('Process took %.3f second' % (t_end - t_start))
```

```python
#check
che_query= """
        select Hvfhs_license_num , count(*) as count_row
        from sample_month_01
        group by Hvfhs_license_num
        order by count_row desc;
"""
pd.read_sql(che_query,engine)
```

```python
#Do pervious process to all data

for data in data_list[1:]: #drop first month because i used it
    print(f"{data}")
    t_start=time()
    re_query = text(f"""
        UPDATE {data}
        SET Hvfhs_license_num = CASE
            WHEN Hvfhs_license_num = 'HV0002' THEN 'Juno'
            WHEN Hvfhs_license_num = 'HV0003' THEN 'Uber'
            WHEN Hvfhs_license_num = 'HV0004' THEN 'Via'
            WHEN Hvfhs_license_num = 'HV0005' THEN 'Lyft'
            ELSE Hvfhs_license_num  -- Keep the existing value if no match
        END;
    """)
    with engine.connect() as conn:
        conn.execute(re_query)
    t_end=time()
    print('Process took %.3f second' % (t_end - t_start))
```
